```python
import pandas as pd
import calendar

data =  pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv', sep=',')
data
```

```python
data.shape
```

```python
data.columns = data.columns.str.upper().str.replace(" ", "_")
data
```

```python
data.drop(columns=data.columns[0],axis=1,inplace=True)
display(data)
```

```python
data.dtypes
```

```python
data.isna().sum()
```

```python
data['RESPONSE'] = data['RESPONSE'].fillna('No')
display(data)
```

```python
display(data["MONTHS_SINCE_LAST_CLAIM"])
```

```python
#when there are no claims or open complaints we set to 0
data['MONTHS_SINCE_LAST_CLAIM'] = data['MONTHS_SINCE_LAST_CLAIM'].fillna(0)
```

```python
data['NUMBER_OF_OPEN_COMPLAINTS'] = data['NUMBER_OF_OPEN_COMPLAINTS'].fillna(0)
```

```python
data['VEHICLE_CLASS'] = data['VEHICLE_CLASS'].fillna('No Vehicle Class')
```

```python
data['VEHICLE_SIZE'] = data['VEHICLE_SIZE'].fillna('no vehicle size')
```

```python
data['VEHICLE_TYPE'] = data['VEHICLE_TYPE'].fillna('no given')
```

```python
data['STATE'] = data['STATE'].fillna('no given')
```

```python
display(data)
```

```python
data.isna().sum()
```

```python
display(data['EFFECTIVE_TO_DATE'])
```

```python
data['EFFECTIVE_TO_DATE'] = pd.to_datetime(data['EFFECTIVE_TO_DATE'],format='%m%d%y')
data['MONTH']=data['EFFECTIVE_TO_DATE'].apply(lambda date:calendar.month_abbr[date.month])
display(data['MONTH'])
```

```python
only_last_quarter = data[(data['MONTH'] == "Feb") | (data['MONTH'] == "Jan") | (data['MONTH'] =="Mar")]
display(only_last_quarter)
```
